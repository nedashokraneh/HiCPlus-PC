```python
import numpy as np
low_res_frames = np.load("../data/divided-data/GM12878_primary/10kb_resolution/chr18-22(down16)(rep2).npy")
enhanced_frames = np.load("../data/enhanced-data/GM12878_primary_enhanced10kb(from16)(rep2)_chr18-22.npy")
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
fig, axs = plt.subplots(16,2,figsize=(100,100), constrained_layout=True)
#plt.subplots_adjust(hspace = .1, wspace=.01)
axs = axs.ravel()
for i in range(16):
    axs[2*i].matshow(low_res_frames[i,6:34,6:34], cmap='YlOrRd')
    axs[2*i+1].matshow(enhanced_frames[i,:,:], cmap='YlOrRd')
    
```
